# Yahoo Finance Demo / Yahoo Finance 数据示例

这个 notebook 从 Yahoo Finance 下载一组多资产市场数据，比较价格路径、收益率和近期波动。  
This notebook downloads a small multi-asset panel from Yahoo Finance and compares price paths, returns, and recent volatility.

示例资产包括：  
Sample tickers include:

- `SPY`
- `TLT`
- `GLD`
- `UUP`
- `^VIX`


```python
import importlib.util

# 只检查 yfinance 依赖，不在 notebook 内自动安装，避免 Windows/Jupyter 子进程编码问题。
# Only check for yfinance instead of auto-installing it inside the notebook, which avoids Windows/Jupyter subprocess encoding issues.
if importlib.util.find_spec("yfinance") is None:
    raise ImportError("Missing package: yfinance. 请先安装：pip install yfinance")

import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
from IPython.display import display

plt.style.use("ggplot")

TICKERS = {
    "SPY": "US equities / 美国股票",
    "TLT": "Long Treasuries / 长久期美债",
    "GLD": "Gold / 黄金",
    "UUP": "US dollar ETF proxy / 美元 ETF 代理",
    "^VIX": "Volatility index / 波动率指数",
}
```

```python
# 下载市场数据，并快速看价格与收益率摘要。
# Download market data and inspect basic price and return summaries.
raw = yf.download(
    tickers=list(TICKERS),
    start="2015-01-01",
    auto_adjust=True,
    progress=False,
)

close = raw["Close"].copy().dropna(how="all")
returns = close.pct_change().dropna()

display(close.tail())
display(returns.describe().T[["mean", "std", "min", "max"]])
```

```python
# 一张图看相对价格走势，另一张图看近 21 日滚动波动率。
# Use one chart for relative price paths and another for 21-day rolling volatility.
normalized = close.div(close.iloc[0]).mul(100)
rolling_vol = returns.rolling(21).std().mul(100)

fig, axes = plt.subplots(2, 1, figsize=(12, 10))

normalized.plot(ax=axes[0], linewidth=2)
axes[0].set_title("Normalized price paths (start = 100) / 标准化价格路径")
axes[0].set_ylabel("Index level / 指数水平")

rolling_vol.plot(ax=axes[1], linewidth=1.6)
axes[1].set_title("21-day rolling volatility (%) / 21 日滚动波动率")
axes[1].set_ylabel("Percent / 百分比")

plt.tight_layout()
plt.show()
```

## Why Yahoo is useful / 为什么 Yahoo 有用

Yahoo Finance 更像“市场交易数据源”，而不是“宏观指标库”。  
Yahoo Finance is primarily a market-trading data source rather than a macro indicator library.

它适合回答这些问题：  
It is useful for questions like:

- 哪类资产最近更强或更弱。  
  Which assets have recently been stronger or weaker.
- 市场波动是不是在抬升。  
  Whether market volatility is rising.
- 股票、债券、黄金、美元这些资产之间有没有联动变化。  
  Whether stocks, bonds, gold, and the dollar are changing their co-movement.

这也是它在 Track A 里最有价值的地方。  
That is exactly why it is the most immediately useful source for Track A.
